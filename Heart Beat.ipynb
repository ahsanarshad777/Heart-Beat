```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dataFrame = pd.read_csv ('mitbih_train.csv', header = None)
```

```python
dataFrame
```

```python
dataFrame.isnull().sum()
```

```python
dataFrame = dataFrame.fillna(np.mean(dataFrame))
```

```python
dataFrame.isnull().sum()
```

```python
x = dataFrame.drop([187], axis = 1)
y = dataFrame[187]
```

```python
type(x)
```

```python
plt.plot(x[20])
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Heart Beat')
plt.show()
```

```python
x.info()
```

```python
x.describe()
```

```python
x.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, shuffle = True , test_size = 0.5)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(10)
```

```python
knn.fit(X_train , y_train)
```

```python
prediction = knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(prediction, y_test)
```
